```python


# 定义 API 密钥
API_KEY = 'AIzaSyAx8tnQZ-PzfSxCQdQkbmfMXVr-JbF2PCM'

# 定义请求头
headers = {
  'Content-Type': 'application/json',
}

# 定义请求正文
body = {
  "contents": [
    {
      "parts": [
        {
          "text": "Write a story about a magic backpack"
        }
      ]
    }
  ]
}

# 发送请求
response = requests.post(
  'https://gemini.cghdfyyyw.work/v1beta/models/gemini-pro:generateContent?key={}'.format(API_KEY),
  headers=headers,
  json=body,
)

# 检查响应状态码
if response.status_code == 200:
  # 解析响应数据
  data = response.json()

  # 打印响应内容
  for candidate in data['candidates']:
    print(candidate['content']['parts'])
else:
  # 打印错误信息
  print('Error:', response.status_code, response.text)
```

```python
import base64

# 将图片转换为 base64 编码
with open("/new_data/gehui/projects/audio4video/pictures/tokyo/00000001.jpg", "rb") as f:
    image_data = f.read()
image_base64 = base64.b64encode(image_data).decode("utf-8")

# 将 base64 编码转换为字符串，不包含换行符
image_base64_str = image_base64.replace("\n", "")

# 打印 base64 编码
print(image_base64_str)
```

```python
body = {
  "contents":[
    {
      "parts":[
        {"text": "describe the video based on the picture captured from it"},
        {
          "inline_data": {
            "mime_type":"image/jpeg",
            "data": f'{image_base64_str}'
          }
        }
      ]
    }
  ]
}
response = requests.post(
  'https://gemini.cghdfyyyw.work/v1beta/models/gemini-pro-vision:generateContent?key={}'.format(API_KEY),
  headers=headers,
  json=body,
)

# 检查响应状态码
if response.status_code == 200:
  # 解析响应数据
  data = response.json()

  # 打印响应内容
  for candidate in data['candidates']:
    print(candidate['content']['parts'][0]['text'])
else:
  # 打印错误信息
  print('Error:', response.status_code, response.text)
```
